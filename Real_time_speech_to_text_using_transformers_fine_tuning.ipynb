```python
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition", model = "faraahahaha/fine-tune-Wav2Vec2-XLS-R-300M-Indonesia")
```

```python
!pip install gradio
```

```python
import gradio as gr
```

```python
def transcribe(audio, state=""):
    text = p(audio)["text"]
    state += text + " "
    return state, state

# Set the starting state to an empty string

gr.Interface(
    fn=transcribe, 
    inputs=[
        gr.Audio(source="microphone", type="filepath", streaming=True), 
        "state" 
    ],
    outputs=[
        "textbox",
        "state"
    ],
    live=True).launch(debug=True)
```
