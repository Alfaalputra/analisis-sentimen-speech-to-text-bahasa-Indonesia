```python
import itertools
```

```python
uniq = []
data = cp
```

```python
def getUnique(data):
      # Return  list dengan unique names of features
        dif_atrib = []
        n_columns = len(data.columns)
        for column in range(n_columns):
            dif_atrib.extend(list(data.iloc[:, column].unique()))
        
        uniq = list(set(dif_atrib))
        return uniq

uniq = getUnique(data)
uniq
```

```python
def makeTable(data, verbose = True):
        columns_table = uniq 
        if np.nan in columns_table:
            columns_table.remove(np.nan)
        elif 'nan' in columns_table:
            columns_table.remove('nan')
        elif 'NaN' in columns_table:
            columns_table.remove('NaN')
        getUnique(data)
        
        dict_index = {}
        lst_index = []

        if verbose:
            for column in columns_table:
                for i in range(len(data.columns)):
                    lst_index.extend(list(data.loc[data[i] == column].index.values))
                    if i == len(data.columns) - 1 :
                        dict_index[column] = list(set(lst_index))
                        lst_index = []
        else:
            for column in columns_table:
                for i in range(len(data.columns)):
                    lst_index.extend(list(data.loc[data[i] == column].index.values))
                    if i == len(data.columns) - 1 :
                        dict_index[column] = list(set(lst_index))
                        lst_index = []
        
        data_init = {}
        
        for i in columns_table:
            data_init[i] = [0 for i in range(len(data))]
            
        df_table = pd.DataFrame(data_init)
        
        #mengubah value menjadi 1 saat indeks memenuhi value nama kolom
        for name_column in columns_table:
            df_table.loc[dict_index[name_column], name_column] = 1
        
        return df_table

df_bin = makeTable(data)
df_bin
```

```python
def support(min_support=None):
        column_names = df_bin.columns.values
        support = {}
        support_min = {}
        total = len(df_bin)
        if min_support == None:
            for column in column_names:
                numerator = df_bin.loc[df_bin[column] != 0, column].sum()
                support[column] = numerator / total
            return support
        
        else:
            min_support = float(min_support)
            
            for column in column_names:
                numerator = df_bin.loc[df_bin[column] != 0, column].sum()
                support[column] = numerator / total
            
            for key, value in support.items():
                if value >= min_support:
                    support_min[key] = value
                    
            return support_min 

sup = support()
sup
```

```python
def makeQuery(lst=[]):
        str_query = ''
        for item in lst:
            if item == lst[-1]:
                str_query = ''.join(str_query+'`{}` == 1'.format(item))
                break
            str_query = ''.join(str_query+'`{}` == 1 and '.format(item))
        return (str_query)

quer = makeQuery()
quer
```

```python
def fit_all(df_bin, min_support=0.5, separator=' & ', verbose= True, min_combination=1):
        if min_support == None:
            min_support = 0
            
        support_dict = sup
        
        total = len(data)
        dict_finally_support = {}
        dict_finally_index = {}
        test_support = []
        if verbose:
            #permutasi fitur berdasarkan angka dari total fitur permutasi
            for j in range(min_combination, len(df_bin.columns) + 1):
                print('Combination {} by {}'.format(j, j))
                          
                for i in itertools.combinations(support_dict.keys(), r=j): #permutasi
                    get_query = makeQuery(list(i)) #get string dengan kondisi untuk query
                    
                    try:
                        numerator = len(df_bin.query('{}'.format(get_query)).loc[:, list(i)])
                    except:
                        continue
                    #hitung nilai support
                    support = numerator / total
                    
                    
                    #Filter features < min_support
                    if support < min_support:
                        continue

                    test_support.append(support)
                    #create dictionary with answers
                    dict_finally_support[separator.join(list(i))] = support
                    dict_finally_index[separator.join(list(i))] = list(df_bin.query('{}'.format(get_query)).loc[:, list(i)].index)
                
                #berhenti saat support = 0 atau hanya satu nilai support
                if test_support.count(0) == len(test_support) - 1 or any(test_support) == False:
                    print('Stopping in combination {} by {}. All supports = 0.'.format(j, j))
                    break
                test_support = []
            return dict_finally_index, dict_finally_support
        
        else:
            for j in range(2, len(df_bin.columns) + 1):
                for i in itertools.combinations(support_dict.keys(), r=j):
                    get_query = makeQuery(list(i))
                    try:
                        numerator = len(df_bin.query('{}'.format(get_query)).loc[:, list(i)])
                    except:
                        continue
                    support = numerator / total
                    
                    if support < min_support:
                        continue
                    
                    test_support.append(support)
                    dict_finally_support[separator.join(list(i))] = support
                    dict_finally_index[separator.join(list(i))] = list(df_bin.query('{}'.format(get_query)).loc[:, list(i)].index)
                
                if test_support.count(0) == len(test_support) - 1 or any(test_support) == False:
                    break
                test_support = []
            return dict_finally_index, dict_finally_support

fit = fit_all(df_bin)
fit
```

```python
def fit(data, df_bin, min_support=0.5, min_combination = 1, max_combination = 5,separator=' & ', verbose=True):
        if min_support == None:
            min_support = 0
            
        support_dict = sup
        
        total = len(data)
        dict_finally_support = {}
        dict_finally_index = {}
        if verbose:
            for j in range(min_combination, max_combination + 1): #min combination sampai max combination
                print('Combination {} by {}'.format(j, j))
                for i in itertools.combinations(support_dict.keys(), r=j):                    
                    get_query = makeQuery(list(i))
                    
                    try:
                        numerator = len(df_bin.query('{}'.format(get_query)).loc[:, list(i)])
                    except:
                        continue
                    support = numerator / total
                    
                    if support < min_support:
                        continue
                    
                    dict_finally_support[separator.join(list(i))] = support
                    dict_finally_index[separator.join(list(i))] = list(df_bin.query('{}'.format(get_query)).loc[:, list(i)].index)
            
            
            return dict_finally_index,dict_finally_support
        
        else:
            for j in range(min_combination, max_combination + 1):
                for i in itertools.combinations(support_dict.keys(), r=j):
                    get_query = makeQuery(list(i))
                    
                    try:
                        numerator = len(df_bin.query('{}'.format(get_query)).loc[:, list(i)])
                    except:
                        continue
                    support = numerator / total
                    
                    if support < min_support:
                        continue
                    
                    dict_finally_support[separator.join(list(i))] = support
                    dict_finally_index[separator.join(list(i))] = list(df_bin.query('{}'.format(get_query)).loc[:, list(i)].index)
                               
                
            return dict_finally_index, dict_finally_support

fitt = fit(data, df_bin, max_combination = 5)
fitt
```

```python
indeks_rules, support_rules = fit(data, df_bin, 
                                  min_support = 0.5,
                                  min_combination = 1,
                                  max_combination = 3)
```
